```python
# 5. Handle missing values by imputing them with the mean for numerical columns.
#  Display the updated dataset

import pandas as pd
import numpy as np
numerical_cols = df.select_dtypes(include=np.number).columns
for col in numerical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mean())
print("\nUpdated Dataset (after imputing missing numerical values):")
print(df)
```

```python
#7. Identify features that have a correlation value higher than 0.8
  # or lower than -0.8. Write a script to highlight these pairs of attributes.

import pandas as pd
import numpy as np
numeric_df = df.select_dtypes(include=[np.number])
correlation = numeric_df.corr()
# Identify features with high correlation
threshold = 0.8
high_corr_pairs = []

for i in range(len(correlation.columns)):
    for j in range(i + 1, len(correlation.columns)):
        col1 = correlation.columns[i]
        col2 = correlation.columns[j]
        corr_val = correlation.iloc[i, j]

        if abs(corr_val) > threshold:
            high_corr_pairs.append((col1, col2, corr_val))

# Print the highly correlated feature pairs
print("\nHighly Correlated Feature Pairs:")
for pair in high_corr_pairs:
    print(f"{pair[0]} and {pair[1]}: {pair[2]}")
```

```python
# prompt: Plot histograms for numerical attributes:

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Japan_life_expectancy.csv")
numerical_cols = df.select_dtypes(include=['number']).columns
if not numerical_cols.empty:
    # Iterate over numerical columns and plot histograms
    for col in numerical_cols:
        plt.figure(figsize=(4, 2))
        sns.histplot(df[col], kde=True) # Added kde for better visualization
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
else:
    print("No numerical columns found in the dataframe.")
```

```python
# prompt: explain "sns.histplot(df[col], kde=True) # Added kde for better visualization"

The line `sns.histplot(df[col], kde=True)` creates a histogram of the numerical column specified by `col` in the DataFrame `df`, and overlays a Kernel Density Estimate (KDE) plot on top of it. Let's break down each part:

* **`sns.histplot(...)`**: This is the core function call from the Seaborn library (`sns`) to generate a histogram.  Histograms are a visual representation of the distribution of numerical data. They divide the data into bins and show the number of data points falling into each bin.

* **`df[col]`**: This selects a single column from the DataFrame `df`.  The `col` variable represents the name of the numerical column being plotted (e.g., 'Year', 'Life_expectancy').

* **`kde=True`**:  This is the crucial part of the line.  It instructs `histplot` to also plot a KDE.  A KDE is a way to estimate the probability density function of a continuous random variable.  It smooths out the histogram, creating a curve that represents the underlying distribution of the data more clearly.  The KDE helps to visualize the overall shape of the distribution, especially when the data is somewhat sparse or the binning of the histogram obscures the true shape.


**Why use a KDE?**

* **Smoothing:** Histograms can appear jagged or blocky because of the discrete bins. The KDE provides a smoother representation of the distribution.

* **Visualization of Density:** The height of the KDE curve at any point is proportional to the probability density of the data at that point.  This allows for a more precise visualization of where the data is most concentrated.

* **Identifying the Shape of the Distribution:**  KDEs are excellent for visually determining if the data is normally distributed, skewed, or follows some other pattern.  This is useful for statistical analysis and model selection.

In summary, the line adds a KDE to the histogram, enhancing its visual clarity and providing a more accurate representation of the data's distribution.
```

```python
#  Generate box plots for multiple numerical attributes in a single
# plot to compare their distributions. Write the code and describe the
# differences in spread and central tendencies.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Japan_life_expectancy.csv")
numerical_cols = df.select_dtypes(include=['number']).columns
if not numerical_cols.empty:
    # Create a box plot for all numerical attributes.
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df[numerical_cols])
    plt.xticks(rotation=45,ha='right')  # Rotate x-axis labels for better readability
    plt.title('Box Plots of Numerical Attributes')
    plt.show()

    # Describe the differences in spread and central tendencies.
    print("Analysis of Spread and Central Tendency:")
    for col in numerical_cols:
        median = df[col].median()
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        print(f"\nAttribute: {col}")
        print(f"  Median: {median}")
        print(f"  Interquartile Range (IQR): {iqr}")
else:
    print("No numerical columns found in the DataFrame.")
```

```python
# prompt: ha='right' explain its purpose

The line `plt.xticks(rotation=45, ha='right')` is used to rotate the x-axis tick labels in a matplotlib plot. Let's break down its components:

* **`plt.xticks(...)`**: This is a matplotlib function that controls the appearance of the x-axis ticks and their labels.

* **`rotation=45`**: This argument rotates the x-axis tick labels by 45 degrees.  This is particularly helpful when the labels are long or overlap, making the plot more readable.

* **`ha='right'`**: This argument sets the horizontal alignment of the text for the x-axis tick labels. `ha='right'` aligns the text to the right edge of the tick mark.  This is often used in conjunction with rotation so that the labels are readable even when rotated.  Without this alignment, rotated labels might overlap with each other or with the plot itself.


**In the context of the code:**

The code generates box plots for multiple numerical attributes.  If the names of the numerical columns are long, rotating the labels 45 degrees prevents them from overlapping.  The `ha='right'` ensures the labels are readable after rotation because they're aligned with the right edge.
```

```python
#  Write a script to create a box plot for a numerical attribute and
# identify the outliers. What thresholds define these outliers?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Japan_life_expectancy.csv")
def identify_outliers(df, numerical_attribute):
    plt.figure(figsize=(4, 2))
    sns.boxplot(x=df[numerical_attribute])
    plt.title(f'Box Plot of {numerical_attribute}')
    plt.xlabel(numerical_attribute)
    plt.show()
    Q1 = df[numerical_attribute].quantile(0.25)
    Q3 = df[numerical_attribute].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[numerical_attribute] < lower_bound) | (df[numerical_attribute] > upper_bound)]

    print(f"Outliers in {numerical_attribute}:")
    print(outliers)
    print(f"\nOutlier thresholds:")
    print(f"Lower bound: {lower_bound}")
    print(f"Upper bound: {upper_bound}")
identify_outliers(df, 'Life_expectancy')
```

```python
# Write a script to overlay another dataset (e.g., population
# density or income levels) on the existing choropleth map to provide
# more context for the data being visualized. How does the additional data
# change the insights from the map?

# !pip install geopandas
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

try:
    japan_map = gpd.read_file("japan.geojson")
    life_expectancy_data = pd.read_csv("Japan_life_expectancy.csv")
    merged_data = japan_map.merge(life_expectancy_data, left_on='nam', right_on='Prefecture', how='left')
    fig, ax = plt.subplots(1, 1, figsize=(4, 4))
    merged_data.plot(column='Life_expectancy', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)  # Color based on life expectancy
    plt.title("Japan Life Expectancy with Overlay Data")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.show()
except FileNotFoundError:
    print("Error: One or both of the input files were not found.")
except KeyError as e:
    print(f"Error: A key error occurred. Check for the correct key: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
#UPPER TRIANGULAR MATRIX
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Japan_life_expectancy.csv")
plt.figure(figsize=(18, 6))
numeric_df = df.select_dtypes(include=[np.number])
correlation = numeric_df.corr()

# Mask for the upper triangle
mask = np.triu(np.ones_like(correlation, dtype=bool))

# Plot the heatmap
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5, mask=mask)
plt.title('Correlation Heatmap (Upper Triangular)')
plt.show()
```
